## **Importing the Libraries**

```python
import cv2
from PIL import Image
from IPython.display import display
#from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
```

## **Loading the Image**

```python
img = cv2.imread("/content/blue butterfly.jpg")
img = cv2.resize(img, (600, 500))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## **Converting Image to Gray Scale**

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

## **Inverting the Image**

```python
invert_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(invert_image))
```

## **We will be smoothing the image to obtain less edgy and smooth image**

```python
blur_image = cv2.GaussianBlur(invert_image, (5, 5), 0, 0)
display(Image.fromarray(blur_image))
```

## **Final Result - Pencil Sketch**

```python
invert_blur_image = 255 - blur_image
pencil_sketch = cv2.divide(gray_image, invert_blur_image, scale=250.0)
display(Image.fromarray(pencil_sketch))
```

## **Pencil Sketch & Original Image**

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
display(Image.fromarray(pencil_sketch))
```
